```python
def correlation(dataset, threshold):
    col_corr = set()
    corr_matrix = dataset.corr()
    for i in range(len(corr_matrix.columns)):
        for j in range(i):
            if abs(corr_matrix.iloc[i, j]) > threshold: 
                colname = corr_matrix.columns[i]
                col_corr.add(colname)
    return col_corr
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generate synthetic data (for example)
np.random.seed(42)
X = 2 * np.random.rand(100, 1)
y = 4 + 3 * X + np.random.randn(100, 1)

# Define the cost function (Mean Squared Error)
def cost_function(theta0, theta1, X, y):
    m = len(y)
    predictions = theta0 + theta1 * X
    return (1 / (2 * m)) * np.sum((predictions - y) ** 2)

# Generate theta0 and theta1 values
theta0_vals = np.linspace(0, 6, 50)
theta1_vals = np.linspace(0, 6, 50)
theta0, theta1 = np.meshgrid(theta0_vals, theta1_vals)

# Compute cost values
cost_vals = np.array([
    cost_function(t0, t1, X, y) for t0, t1 in zip(np.ravel(theta0), np.ravel(theta1))
])
cost_vals = cost_vals.reshape(theta0.shape)

# Plot the surface
fig = plt.figure(figsize=(10, 6))
ax = fig.add_subplot(111,projection='3d')
ax.plot_surface(theta0, theta1, cost_vals, cmap='viridis', alpha=0.8)
ax.set_title("Cost Function Surface")
ax.set_xlabel(r"$\theta_0$")
ax.set_ylabel(r"$\theta_1$")
ax.set_zlabel("Cost")
plt.show()
```

## Feature Scaling and Standard Scaling
