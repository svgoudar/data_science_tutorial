```python
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats

# Load the dataset
networks = sns.load_dataset('brain_networks', header=[0, 1, 2], index_col=1)

# Select a specific brain region's data
region_data = networks.iloc[:, 10]  # Arbitrary selection of a region

# Create a QQ plot
plt.figure(figsize=(8, 6))
stats.probplot(region_data.dropna(), dist="norm", plot=plt)

# Add title and labels
plt.title('QQ Plot of Brain Region Data (Region 10)')
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the Iris dataset
iris = sns.load_dataset('iris')

# Set the theme to 'ggplot' style
# plt.style.use('ggplot')

# Create a scatterplot of petal length vs. petal width, colored by species
sns.scatterplot(data=iris, x='petal_length', y='petal_width', hue='species', palette='Set1')

# Add titles and labels
plt.title("Petal Length vs Petal Width by Species")
plt.xlabel("Petal Length (cm)")
plt.ylabel("Petal Width (cm)")

# Show the plot
plt.show()
```
