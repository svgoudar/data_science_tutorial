```python
import numpy as np
import matplotlib.pyplot as plt

# Step 1: Generate some random 2D data points
np.random.seed(42)
data = np.random.randn(100, 2)

# Step 2: Standardize the data (zero-centering)
mean_data = np.mean(data, axis=0)
print(mean_data)
std_data = data - mean_data
print("Sanjeev : ",std_data)
print("Sanjeev : ",std_data.T)
# Step 3: Calculate the covariance matrix of the standardized data
cov_matrix = np.cov(std_data.T)
print("Covariance Matrix:")
print(cov_matrix)

# Step 4: Perform eigen decomposition of the covariance matrix
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
print("\nEigenvalues:", eigenvalues)
print("Eigenvectors:\n", eigenvectors)

# Step 5: Sort eigenvalues and eigenvectors in descending order
sorted_indices = np.argsort(eigenvalues)[::-1]
sorted_eigenvalues = eigenvalues[sorted_indices]
sorted_eigenvectors = eigenvectors[:, sorted_indices]

# Step 6: Principal Component Analysis (PCA)
principal_component1 = sorted_eigenvectors[:, 0]
principal_component2 = sorted_eigenvectors[:, 1]

# Plotting the data and principal components
plt.scatter(std_data[:, 0], std_data[:, 1], alpha=0.5)
origin = [0, 0]  # origin point for the eigenvectors

# Plot eigenvectors (principal components)
plt.quiver(*origin, principal_component1[0], principal_component1[1], color=['r'], scale=3, label="PC1")
plt.quiver(*origin, principal_component2[0], principal_component2[1], color=['b'], scale=3, label="PC2")

plt.axhline(0, color='black',linewidth=0.5)
plt.axvline(0, color='black',linewidth=0.5)
plt.grid()
plt.title("Data Points with Principal Components")
plt.legend()
plt.show()
```
